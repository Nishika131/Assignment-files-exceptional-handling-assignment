9)What is the difference between multithreading and multiprocessing

Difference Between Multithreading and Multiprocessing in Python
Feature	Multithreading	Multiprocessing
Definition	Multiple threads within a single process	Multiple independent processes
Performance	Better for I/O-bound tasks (e.g., file, network)	Better for CPU-bound tasks (e.g., heavy computation)
GIL Impact	Affected by GIL (Global Interpreter Lock)	Not affected — each process has its own Python interpreter
Memory Use	Shared memory	Separate memory space

```python
#practical questions

# 1. Open a file for writing and write a string to it
with open("output.txt", "w") as file:
    file.write("Hello, this is a test message.")
print("1. String written to output.txt")

# 2. Read the contents of a file and print each line
with open("output.txt", "r") as file:
    print("2. Contents of output.txt:")
    for line in file:
        print(line.strip())

# 3. Handle file not found error
try:
    with open("non_existent.txt", "r") as file:
        print(file.read())
except FileNotFoundError:
    print("3. The file does not exist.")

# 4. Copy content from one file to another
with open("output.txt", "r") as source, open("copied.txt", "w") as dest:
    for line in source:
        dest.write(line)
print("4. Contents copied to copied.txt")

# 5. Handle division by zero
try:
    x = 10 / 0
except ZeroDivisionError:
    print("5. Cannot divide by zero.")

# 6. Log an error message on ZeroDivisionError
import logging
logging.basicConfig(filename="app.log", level=logging.ERROR)
try:
    result = 10 / 0
except ZeroDivisionError:
    logging.error("6. Division by zero occurred.")
    print("6. Error logged to app.log")

# 7. Log different levels
logging.basicConfig(level=logging.DEBUG)
logging.info("7. This is an info message.")
logging.warning("7. This is a warning.")
logging.error("7. This is an error.")
print("7. Logged messages of INFO, WARNING, and ERROR levels")

# 8. Handle file opening error
try:
    file = open("data.txt", "r")
except FileNotFoundError:
    print("8. File not found!")
else:
    print(file.read())
    file.close()

# 9. Read file lines into a list
with open("output.txt", "r") as file:
    lines = file.readlines()
print("9. Lines as list:", lines)

# 10. Append data to a file
with open("log.txt", "a") as file:
    file.write("10. Appended log entry\n")
print("10. Appended to log.txt")

# 11. Handle KeyError in dictionary
my_dict = {"name": "Nishika"}
try:
    print(my_dict["age"])
except KeyError:
    print("11. Key 'age' not found in dictionary.")

# 12. Multiple except blocks
try:
    num = int("abc")
    result = 10 / num
except ValueError:
    print("12. Invalid input.")
except ZeroDivisionError:
    print("12. Division by zero.")

# 13. Check if a file exists
import os
if os.path.exists("output.txt"):
    with open("output.txt", "r") as file:
        print("13. File exists. Contents:", file.read())
else:
    print("13. File does not exist.")

# 14. Log info and error
logging.basicConfig(filename="info_error.log", level=logging.INFO)
try:
    result = 10 / 0
except ZeroDivisionError:
    logging.error("14. Division by zero error.")
else:
    logging.info("14. Division successful.")
print("14. Logged info/error to info_error.log")

# 15. Handle empty file
with open("empty.txt", "w"):  # Create empty file
    pass
with open("empty.txt", "r") as file:
    content = file.read()
    if content:
        print(content)
    else:
        print("15. The file is empty.")

# 16. Memory profiling example (install memory_profiler manually in Colab)
# !pip install memory_profiler
from memory_profiler import profile
@profile
def create_list():
    return [i for i in range(10000)]
create_list()
print("16. Memory profiling complete")

# 17. Write list of numbers to file
numbers = [1, 2, 3, 4, 5]
with open("numbers.txt", "w") as file:
    for num in numbers:
        file.write(f"{num}\n")
print("17. Written numbers to numbers.txt")

# 18. Logging with rotation
from logging.handlers import RotatingFileHandler
handler = RotatingFileHandler("rotating_log.log", maxBytes=1048576, backupCount=3)
logging.basicConfig(handlers=[handler], level=logging.INFO)
logging.info("18. Test log message with rotation")
print("18. Logged with rotation")

# 19. Handle IndexError and KeyError
try:
    my_list = [1, 2, 3]
    print(my_list[5])
    my_dict = {"a": 1}
    print(my_dict["b"])
except IndexError:
    print("19. Index out of range.")
except KeyError:
    print("19. Key not found.")

# 20. Read file using context manager
with open("output.txt", "r") as file:
    content = file.read()
    print("20. File content:", content)

# 21. Count occurrences of a word
with open("word_count.txt", "w") as f:
    f.write("Python is great. Python is easy. Python!")
word_to_find = "Python"
count = 0
with open("word_count.txt", "r") as file:
    for line in file:
        count += line.count(word_to_find)
print(f"21. '{word_to_find}' occurred {count} times.")

# 22. Check if file is empty
if os.stat("empty.txt").st_size == 0:
    print("22. File is empty.")
else:
    with open("empty.txt", "r") as file:
        print(file.read())

# 23. Log error during file handling
logging.basicConfig(filename="file_error.log", level=logging.ERROR)
try:
    with open("non_existing_file.txt", "r") as file:
        print(file.read())
except FileNotFoundError as e:
    logging.error(f"23. File handling error: {e}")
    print("23. Error logged to file_error.log")
```
